```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path='Visadataset.csv'
pd.read_csv(path)
```

```python
visa_df=pd.read_csv(path)
visa_df
```

```python
q1=np.percentile(visa_df['prevailing_wage'],25)
q2=np.percentile(visa_df['prevailing_wage'],50)
q3=np.percentile(visa_df['prevailing_wage'],75)
iqr=q1-q3
lb=q1-1.5*iqr
ub=q3-1.5*iqr
c1=visa_df['prevailing_wage']<lb
c2=visa_df['prevailing_wage']>ub
condition=c1|c2
outliers_df=visa_df[condition]
print("------------------Outliers dataframe_____________________________\n")
outliers_df
```

```python
list1=[]
q1=np.percentile(visa_df['prevailing_wage'],25)
q2=np.percentile(visa_df['prevailing_wage'],50)
q3=np.percentile(visa_df['prevailing_wage'],75)
iqr=q1-q3
lb=q1-1.5*iqr
ub=q3+1.5*iqr
median=visa_df['prevailing_wage'].median()
for i in visa_df['prevailing_wage']:
    if i<lb or i>ub:
        list1.append(median)
    else:
        list1.append(i)
visa_df['prevailing_wage_new']=list1
visa_df['prevailing_wage_new']
```

```python
plt.boxplot(visa_df['prevailing_wage_new'])
plt.show()
```

```python
#np.where takes 3 argu con,true,false value
d={}
q1=np.percentile(visa_df['prevailing_wage'],25)
q2=np.percentile(visa_df['prevailing_wage'],50)
q3=np.percentile(visa_df['prevailing_wage'],75)
iqr=q1-q3
lb=q1-1.5*iqr
ub=q3-1.5*iqr
median1=visa_df['prevailing_wage'].median()
c1=visa_df['prevailing_wage']<lb
c2=visa_df['prevailing_wage']>ub
cond=c1|c2
t=median1
f=visa_df['prevailing_wage']
visa_df['prevailing_wage_npwhere']=np.where(cond,t,f)
visa_df['prevailing_wage_npwhere']
```

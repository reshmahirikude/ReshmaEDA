```python
#create list to create DataFrame
name=['Ramesh','Santosh','Mahesh']
age=[30,40,35]
name,age
```

```python
#Save list to the DataFrame
pd.DataFrame(zip(name,age))
```

```python
#Add Column Name to Dataframe
data=zip(name,age)
cols=['Name','Age']
pd.DataFrame(data,columns=cols)
```

```python
#Direct Value Passing to the Dataframe Parameters
#data=zip(name,age)
#cols=['Name','Age']
pd.DataFrame(zip(name,age),columns=['Name','Age'])
```

```python
#Passing index to the Rows
ind=['A','B','C']
pd.DataFrame(zip(name,age),columns=['Name','Age'],index=ind)
```

```python
#Adding New Column to the Dataframe
city_nm=['Hyd','pune','Begaluru']
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
df['City']=city_nm
df
```

```python
#Updateting the Dataframe with existing Data
ind=['A','B','C']
df=pd.DataFrame(zip(name,age),columns=['Name','Age'],index=ind)
df['Name']=['Reshma','Shamali','Mahadevi']
df
```

```python
#Deleting the Dataframe column
city_nm=['Hyd','pune','Begaluru']
data=zip(name,age)
#cols=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=['Name','Age'],index=ind)
df['City']=city_nm
df['Name']=['Reshma','Shamali','Mahadevi']
df.drop('City',axis=1,inplace=True)
#inplace=True Overwrite the dataframe inplace=false create new Dataframe
df
```

```python
#Deleting The Row of the Dataframe make axis=0 to delete row 
#maintion the row index to delete perticular row
city_nm=['Hyd','pune','Begaluru']
ind=['A','B','C']
df1=pd.DataFrame(zip(name,age),columns=['Name','Age'],index=ind)
df1['City']=city_nm
df1['Name']=['Reshma','Shamali','Mahadevi']
df1.drop('C',axis=0,inplace=True)#c is the index of row 
df1
```

```python
#Creating excel file to store df1 dataframe
df1.to_csv("output.csv")
pd.read_csv("output.csv")#read stored Data
#while saving a,b index is saying extra column so you have to give index=false
```

```python
#To avoid  unmaned Column set index as False
df1.to_csv("output.csv",index=False)
pd.read_csv("output.csv")
#while saving a,b index is saying extra column so you have to give index=false
```
